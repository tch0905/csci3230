```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([
    [5.86, 0.74],
    [1.34, 1.18],
    [3.65, 0.51],
    [4.69, -0.48],
    [4.13, -0.07],
    [5.87, 0.37],
    [7.91, 1.35],
    [5.57, 0.30],
    [7.30, 1.64],
    [7.89, 1.75]
])


```

```python
X = data[:, 0]  # Input variables (x)
Y = data[:, 1]  # Output variables (y)
```

```python
X_mean = np.mean(X)
Y_mean = np.mean(Y)
```

```python
# 1(b)
```

```python
temp = np.column_stack((np.ones_like(X),X,X**2))
theta = (np.linalg.inv(temp.T@temp)@temp.T@Y)
print(theta)
```

```python
# Plot the data points
plt.scatter(X, Y, color='blue', label='Data Points')

# Generate x values for the curve
x_curve = np.linspace(min(X), max(X), 100)

# Calculate predicted Y values using the model
Y_pred = theta[0] + theta[1] * x_curve + theta[2] * (x_curve**2)

# Plot the linear model
plt.plot(x_curve, Y_pred, color='red', label='Linear Model')

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Model')

# Add legend
plt.legend()

# Display the plot
plt.show()
```
